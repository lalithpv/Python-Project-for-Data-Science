```python
import yfinance as yf

ticker_symbol = 'GME'
gme = yf.Ticker(ticker_symbol)

historical_data = gme.history(period='1y', interval='1d')  # 1 year of daily data

print(historical_data)

last_close = historical_data['Close'].iloc[-1]
print(f"Last closing price of GameStop (GME): {last_close}")

dividends = gme.dividends
print("Dividends:")
print(dividends)

info = gme.info
print("Company Information:")
print(info)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')

revenue_data = []

if table:
    for row in table.find_all('tr')[1:]:
        columns = row.find_all('td')
        if len(columns) > 1:
            year = columns[0].text.strip()
            revenue = columns[1].text.strip()
            revenue_data.append((year, revenue))

for year, revenue in revenue_data:
    print(f"Year: {year}, Revenue: {revenue}")
```

```python
import yfinance as yf
import pandas as pd

gamestop = yf.Ticker("GME")
stock_data = gamestop.history(period="1y")

print(stock_data.head())
import requests
from bs4 import BeautifulSoup

url = "https://aroussi.com/post/python-yahoo-finance"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

revenue_data = []

table = soup.find('table')

if table:
    for row in table.find_all('tr')[1:]:
        columns = row.find_all('td')
        if len(columns) > 1:
            year = columns[0].text.strip()
            revenue = columns[1].text.strip().replace(',', '').strip()
            revenue_data.append((year, float(revenue)))

revenue_df = pd.DataFrame(revenue_data, columns=['Year', 'Revenue'])
print(revenue_df)
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go

app = dash.Dash(__name__)

stock_data.reset_index(inplace=True)

app.layout = html.Div([
    html.H1("GameStop Stock and Revenue Dashboard"),
    
    dcc.Graph(
        id='stock-price',
        figure={
            'data': [
                go.Scatter(
                    x=stock_data['Date'],
                    y=stock_data['Close'],
                    mode='lines+markers',
                    name='Stock Price'
                )
            ],
            'layout': go.Layout(
                title='GameStop Stock Price',
                xaxis={'title': 'Date'},
                yaxis={'title': 'Stock Price (USD)'},
                hovermode='closest'
            )
        }
    ),
    
    dcc.Graph(
        id='revenue',
        figure={
            'data': [
                go.Bar(
                    x=revenue_df['Year'],
                    y=revenue_df['Revenue'],
                    name='Revenue'
                )
            ],
            'layout': go.Layout(
                title='GameStop Revenue',
                xaxis={'title': 'Year'},
                yaxis={'title': 'Revenue (in billions)'},
                hovermode='closest'
            )
        }
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
